```python
target= clean_data[['survived']]
target
```

```python
training_data = clean_data.drop('survived',axis=1,inplace=False) # inplace=False 원본은 그대로 있고, survived column을 떼어 버리고 값을 리턴시킴. 
training_data.head()
```

```python
value_data = training_data[['age','fare']] 
value_data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data =scaler.fit_transform(value_data)
value_data = pd.DataFrame(scaled_data, columns=value_data.columns)
value_data.head()
```

```python
training_data.drop(['age','fare'],axis=1,inplace=True)
training_data.head()
```

```python
onehot_data = pd.get_dummies(training_data['pclass']) # column을 주면 그것에 맞춰서 늘려줘. 
onehot_data.head()
# ( 밑에 표를 해석하면 ) 1등석이면 1에 True 이고, 아니면 False 처리 됨

#get_dummies 함수는 이러한 가변수화를 쉽게 수행할 수 있는 판다스(Pandas) 라이브러리의 함수 중 하나입니다.
#원핫 인코딩(One-Hot Encoding): 범주형 변수를 이진(0 또는 1) 값으로 이루어진 여러 열로 변환합니다.
#각 범주는 자신의 열에 해당하며, 해당 범주에 속하는 경우 1, 속하지 않는 경우 0으로 표시됩니다.
```

```python
onehot_data = pd.get_dummies(training_data, columns=training_data.columns)
onehot_data.head()
```

```python
# 라벨링을 했기 때문에 0,1로 뜨는 것.
# sex_0	sex_1 
# 라벨링을 안했으면 female, male로 뜰것. 
onehot_data.info()
```

```python
training_data=pd.concat([value_data,onehot_data],axis=1)
training_data.info()
```

```python
training_data.astype('float64')
training_data.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(training_data.astype('float64'), target,test_size=0.20)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()
model.add(Dense(128, input_dim=34, activation='relu'))
model.add(Dropout(0.02))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.02))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.02))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.02))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.02))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.02))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.02))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['binary_accuracy'])
```

```python
fit_hist = model.fit( x_train, y_train,batch_size=50, epochs=5,validation_split=0.2,verbose=1) 
```

```python
import matplotlib.pyplot as plt
plt.plot(fit_hist.history['binary_accuracy'])
plt.plot(fit_hist.history['val_binary_accuracy'])
plt.show()
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('loss', score[0])
print('accuracy',score[1])
```
